```python
# Multiple Linear Regression (Implement Pytorch)

import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim


torch.manual_seed(1) # Fixed random seed


# Training Data

x1_train = torch.FloatTensor([[73],[93],[89],[96],[73]])
x2_train = torch.FloatTensor([[80],[88],[91],[98],[66]])
x3_train = torch.FloatTensor([[75],[93],[90],[100],[70]])
y_train = torch.FloatTensor([[152],[185],[180],[196],[142]])


# Initialize Weight W, bias b

w1 = torch.zeros(1, requires_grad = True) # 1 means the number of 0
w2 = torch.zeros(1, requires_grad = True)
w3 = torch.zeros(1, requires_grad = True)
b = torch.zeros(1, requires_grad = True)


optimizer = optim.SGD([w1, w2, w3, b], lr = 1e-5)   # Set Optimizer 

nb_epochs = 1000

for epoch in range(nb_epochs + 1):

    hypothesis = x1_train * w1 + x2_train * w2 + x3_train * w3 + b   # Calculate H(x)
    cost = torch.mean((hypothesis - y_train) ** 2)   # Calculate Cost

    # Improve H(x) with cost

    optimizer.zero_grad()
    cost.backward()
    optimizer.step()

    # Print Log every 100 times

    if epoch % 100 == 0:
        print('Epoch {:4d} / {} w1: {:.3f} w2: {:.3f} w3: {:.3f} b: {:.3f} Cost: {:.6f}'.format(
            epoch, nb_epochs, w1.item(), w2.item(), w3.item(), b.item(), cost.item()
            ))
```

```python
# Multiple Linear Regression (Implement Pytorch considering matrix operations)


x_train = torch.FloatTensor([[73, 80, 75],
                             [93, 88, 93],
                             [89, 91, 80],
                             [96, 98, 100],
                             [73, 66, 70]])

y_train = torch.FloatTensor([[152], [185], [180], [196], [142]])

print(x_train.shape, y_train.shape)   # x,y shape


# Model initialization


# Declaration of Weight and Deflection

W = torch.zeros((3, 1), requires_grad = True)
b = torch.zeros(1, requires_grad = True)

optimizer = optim.SGD([W, b], lr = 1e-5)   # Set optimizer

nb_epochs = 20

for epoch in range(nb_epochs + 1):

    # Calculate

    hypothesis = x_train.matmul(W) + b   # bias b is broadcasted and added to each sample.
    cost = torch.mean((hypothesis - y_train) ** 2)   # cost

    # Improve H(x) with cost

    optimizer.zero_grad()
    cost.backward()
    optimizer.step()

    print('Epoch {:4d} / {} hypothesis: {} Cost: {:.6f}'.format(
        epoch, nb_epochs, hypothesis.squeeze().detach(), cost.item()
    ))
    
```

```python
# Classify MNIST with CNN


import torch
import torchvision.datasets as dsets
import torchvision.transforms as transforms
import torch.nn.init


device = 'cuda' if torch.cuda.is_available else 'cpu'

torch.manual_seed(777)   # fixed random seed


# Fixed random seeds when GPU is available

if device == 'cuda':
    torch.cuda.manual_seed_all(777)   


# Set parameters

learning_rate = 0.001
training_epochs = 15
batch_size = 100


# Define Data Set

mnist_train = dsets.MNIST(root = 'MNIST_data/', # specify download path
                          train = True, # If True is specified, download as training data
                          transform = transforms.ToTensor(), # convert to tensor
                          download = True)

mnist_test = dsets.MNIST(root = 'MNIST_data/', # specify download path
                         train = False, # If False is specified, download as test data
                         transform = transforms.ToTensor(), # convert to tensor
                         download = True)

data_loader = torch.utils.data.DataLoader(dataset = mnist_train,
                                          batch_size = batch_size,
                                          shuffle = True,
                                          drop_last = True)


class CNN(torch.nn.Module):

    def __init__(self):
        super(CNN, self).__init__()

        # First Layer

        ''' ImgIn shape = (?, 28, 28, 1)
                Conv (?, 28, 28, 32)
                Pool (?, 14, 14, 32) '''
        
        self.layer1 = torch.nn.Sequential(
            torch.nn.Conv2d(1, 32, kernel_size = 3, stride = 1, padding = 1),
            torch.nn.ReLU(),
            torch.nn.MaxPool2d(kernel_size = 2, stride = 2))

        # Second Layer

        ''' ImgIn shape = (?, 14, 14, 32)
                Conv (?, 14, 14, 64)
                Pool (?, 7, 7, 64) '''

        self.layer2 = torch.nn.Sequential(
            torch.nn.Conv2d(32, 64, kernel_size = 3, stride = 1, padding = 1),
            torch.nn.ReLU(),
            torch.nn.MaxPool2d(kernel_size = 2, stride = 2))
        
        # Full-bonded Layer

        self.fc = torch.nn.Linear(7 * 7 * 64, 10, bias = True)   # 7 * 7 * 64, 10 outputs

        # Initialize weights to Full-bonded layer only

        torch.nn.init.xavier_uniform_(self.fc.weight)


    def forward(self, x):

        out = self.layer1(x)
        out = self.layer2(out)
        out = out.view(out.size(0), -1)   # flatten for full-bonded layer
        out = self.fc(out)
        return out
        

model = CNN().to(device)   # define CNN Model 

criterion = torch.nn.CrossEntropyLoss().to(device)   # the cost function contains a softmax function

optimizer = torch.optim.Adam(model.parameters(), lr = learning_rate)

total_batch = len(data_loader)

print('Total batch : {}'.format(total_batch))   # total batch 600 X batch size 100 = total data 60000


for epoch in range(training_epochs):
    
    avg_cost = 0

    for X,Y in data_loader:   # It is taken out in mini-batch units (X : mini batch, Y: label)

        # image is already size of (28 x 28), no reshape
        # label is not one-hot encoded

        X = X.to(device)
        Y = Y.to(device)

        optimizer.zero_grad()
        hypothesis = model(X)
        cost = criterion(hypothesis, Y)
        cost.backward()
        optimizer.step()

        avg_cost += cost / total_batch

    print('[Epoch : {:>4}] cost = {:>.9}'.format(epoch + 1, avg_cost))


with torch.no_grad():

    X_test = mnist_test.test_data.view(len(mnist_test), 1, 28, 28).float().to(device)
    Y_test = mnist_test.test_labels.to(device)

    prediction = model(X_test)
    correct_prediction = torch.argmax(prediction, 1) == Y_test
    accuracy = correct_prediction.float().mean()
    print('Accuracy : ', accuracy.item())
```
